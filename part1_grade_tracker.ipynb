Data Parsing and Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
cleaned_students = []

for student in raw_students:
    #Cleaning name
    name = student["name"].strip().title()
    
    #Converting roll to int
    roll_no = int(student["roll"])
    
    #Converting marks_str to list of integers
    marks = list(map(int, student["marks_str"].split(", ")))
    
    #Validating name (only alphabetic words)
    is_valid = all(word.isalpha() for word in name.split())
    
    #Printing validation
    if is_valid:
        print(f"{name} ✓ Valid name")
    else:
        print(f"{name} ✗ Invalid name")
    
    #Storing cleaned data
    cleaned_students.append({
        "name": name,
        "roll": roll_no,
        "marks": marks
    })
    
    #Printing formatted profile card
    print("=" * 32)
    print(f"Student : {name}")
    print(f"Roll No : {roll_no}")
    print(f"Marks   : {marks}")
    print("=" * 32)

#Finding student with roll number 103
for student in cleaned_students:
    if student["roll"] == 103:
        print("\nSpecial Output for Roll 103:")
        print(student["name"].upper())
        print(student["name"].lower())
```

Subject-wise Grades (Using for loop)

```python


Student_name = "Ayesha Sharma"
Subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
Marks        = [88, 72, 95, 60, 78]

print(f"\nReport Card for {Student_name}\n")

# using for loop

for i in range(len(Subjects)):
    marks = Marks[i]
    
    # Grade logic
    if 90 <= marks <= 100:
        grade = "A+"
    elif 80 <= marks <= 89:
        grade = "A"
    elif 70 <= marks <= 79:
        grade = "B"
    elif 60 <= marks <= 69:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{Subjects[i]:<20}{marks:<10}{grade:<10}")
    print ("="*40)

 
```

Marks Analysis

```python
# Total
total = sum(Marks)

# Average
average = round(total / len(Marks), 2)

# Highest
max_marks = max(Marks)
max_index = Marks.index(max_marks)
highest_subject = Subjects[max_index]

# Lowest
min_marks = min(Marks)
min_index = Marks.index(min_marks)
lowest_subject = Subjects[min_index]

print("\nSummary:")
print(f"Total Marks       : {total}")
print(f"Average Marks     : {average}")
print(f"Highest Scoring   : {highest_subject} ({max_marks})")
print(f"Lowest Scoring    : {lowest_subject} ({min_marks})")
```

Marks Entry (Using while loop)

```python
new_subjects = []
new_marks = []

count = 0

while True:
    subject = input("\nEnter subject name (or type 'done' to stop): ")
    
    if subject.lower() == "done":
        break
    
    marks_input = input(f"Enter marks for {subject} (0-100): ")
    
    # Validate marks
    if not marks_input.isdigit():
        print("Do not crash and do not add invalid entries to the list.")
        continue
    
    marks_value = int(marks_input)
    
    if marks_value < 0 or marks_value > 100:
        print("Do not crash and do not add invalid entries to the list.")
        continue
    
    # Valid entry
    new_subjects.append(subject)
    new_marks.append(marks_value)
    count += 1

    # Combine old + new marks
all_marks = Marks + new_marks

if len(all_marks) > 0:
    updated_avg = round(sum(all_marks) / len(all_marks), 2)
else:
    updated_avg = 0

print("\nFinal Update:")
print(f"New Subjects Added : {count}")
print(f"Updated Average    : {updated_avg}")
```

Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

pass_count = 0
fail_count = 0
total_avg_sum = 0

topper_name = ""
topper_avg = 0

# Table Header
print("=" * 45)
print(f"{'Name':<20} | {'Average':<10} | {'Status':<8}")
print("-" * 45)

for name, Marks in class_data:
    avg = round(sum(Marks) / len(Marks), 2)
    total_avg_sum += avg
    
    # Pass/Fail
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1
    
    # Topper check
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name
    
    # Print row
    print(f"{name:<20} | {avg:<10} | {status:<8}")

print("=" * 45)

# Class average
class_avg = round(total_avg_sum / len(class_data), 2)

# Final Summary
print(f"\nPassed Students : {pass_count}")
print(f"Failed Students : {fail_count}")
print(f"Class Topper    : {topper_name} ({topper_avg})")
print(f"Class Average   : {class_avg}")
```

String Manipulation Utility

```python
# Given essay
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Step 1: Strip whitespace
clean_essay = essay.strip()
print("1. Clean Essay:")
print(clean_essay)

# Step 2: Convert into Title Case
title_case = clean_essay.title()
print("\n2. Title Case:")
print(title_case)

# Step 3: Count "python"
count_python = clean_essay.count("python")
print("\n3. Count of 'python':")
print(count_python)

# Step 4: Replace "python" with "Python 🐍"
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("\n4. Replaced Essay:")
print(replaced_essay)

# Step 5: Split into sentences
sentences = clean_essay.split(". ")
print("\n5. Sentence List:")
print(sentences)

# Step 6: Numbered sentences
print("\n6. Numbered Sentences:")
for i in range(len(sentences)):
    sentence = sentences[i].strip()
    
     # Convert to sentence case
    sentence = sentence.capitalize()
    
    # Ensure sentence ends with '.'
    if not sentence.endswith("."):
        sentence += "."
    
    print(f"{i+1}. {sentence}")
```
